# Importing all the libraries required

```python
from tkinter import *
from PIL import ImageTk,Image
from tkinter import filedialog
import cv2
import numpy as np
from tensorflow.keras.models import load_model
```

# Creates the application

```python
#Create a non sizable 700*700 window
window = Tk()
window.geometry("700x700")
window.title("Age, gender and ethnicity detection")
#Add a icon into the bar
window.iconbitmap('D:/University 3rd year/Artificial Intelligence and Robotics/Assigment/Icon.ico')
window.resizable(0, 0)
width=16
height=2
width1=234
width3=10
width4=17
height1=75
padx=5
pady=5

#Add the logo
logo_filename= 'D:/University 3rd year/Artificial Intelligence and Robotics/Assigment/Logo.png'
img2=Image.open(logo_filename)
img2.thumbnail((200,200))
my_image2 = ImageTk.PhotoImage(img2)
my_image2_label=Label(window,image=my_image2,bg='#0C2D48',width=700, height=250).grid(row=0,column=0,columnspan=2)

#Action to load the models and perform the prediction
def handle_click_pediction():
    face_classifier =cv2.CascadeClassifier(r'C:/Users/Alexandru Neagu/Exercises/Team project/haarcascade_frontalface_default.xml')
    classifier1= load_model(r'C:/Users/Alexandru Neagu/Exercises/AI and Robotics/Checkpoint/Model-Gender.h5')
    classifier2= load_model(r'C:/Users/Alexandru Neagu/Exercises/AI and Robotics/Checkpoint_age/Model-Age.h5')
    classifier3= load_model(r'C:/Users/Alexandru Neagu/Exercises/AI and Robotics/Checkpoint_ethnicity/Model-Ethnicity.h5')
    
    class_labels3=['African', 'Asian', 'Caucasian', 'Indian']
    class_labels2=['Child', 'Middle-age adult', 'Old adult', 'Young adult']
    class_labels1=['Female', 'Male']
    freq1=[0,0]
    freq2=[0,0,0,0]
    freq3=[0,0,0,0]
    
    img=cv2.imread(window.filename)
    for i in range(5):
        #Useing Haar cascade to crop the face out of the images
        faces = face_classifier.detectMultiScale( 
        img,
        scaleFactor = 1.1,
        minNeighbors = 4,
        minSize = (30, 30)
       )
        for (x, y, w, h) in faces:
            cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
            roi_gray = img[y:y+h,x:x+w]
            #This line show if the Haar cascade could detect the face in the picture
            #cv2.imshow('image',roi_gray)
            cropped_img = np.expand_dims(np.expand_dims(cv2.resize(roi_gray, (224, 224)), -1), 0)
    
        
            preds1 = classifier1.predict(cropped_img)
            label1=class_labels1[np.argmax(preds1)]

        
            preds2 = classifier2.predict(cropped_img)
            label2=class_labels2[np.argmax(preds2)]
            
            preds3 = classifier3.predict(cropped_img)
            label3=class_labels3[np.argmax(preds3)]
        
        
        #Createing a list with the freaquances of each prediction
            freq1[class_labels1.index(label1)]+=1
            freq2[class_labels2.index(label2)]+=1
            freq3[class_labels3.index(label3)]+=1
    
    #Deleteing the old result and inserting the new one
    if (text_gender['state'] == DISABLED):
            text_gender['state'] = NORMAL
            
    if (text_ethnicity['state'] == DISABLED):
            text_ethnicity['state'] = NORMAL
            
    if (text_age['state'] == DISABLED):
            text_age['state'] = NORMAL
            
    text_gender.delete("1.0","end")
    text_age.delete("1.0","end")
    text_ethnicity.delete("1.0","end") 
    text_gender.insert(END,class_labels1[freq1.index(max(freq1))])
    text_age.insert(END,class_labels2[freq2.index(max(freq2))])
    text_ethnicity.insert(END,class_labels3[freq3.index(max(freq3))])
    
    
            
    text_age['state'] = DISABLED
    text_ethnicity['state'] = DISABLED
    text_gender['state'] = DISABLED

    
#Createing a frame for each frid of the apication
frame1 = Frame(master=window, width=width1, height=height1, bg="#2E8BC0")
frame1.grid(row=1,column=0,sticky=N+S+E+W)
frame2 = Frame(master=window, width=width1, height=height1, bg="#2E8BC0")
frame2.grid(row=2,column=0,sticky=N+S+E+W)
frame3 = Frame(master=window, width=width1, height=height1, bg="#2E8BC0")
frame3.grid(row=3,column=0,sticky=N+S+E+W)
frame5 = Frame(master=window, width=width1, height=height1, bg="#2E8BC0")
frame5.grid(row=5,column=0,sticky=N+S+E+W)
frame6 = Frame(master=window, width=width1, height=height1, bg="#2E8BC0")
frame6.grid(row=6,column=0,sticky=N+S+E+W)
frame7 = Frame(master=window, width=width1, height=height1, bg="#2E8BC0")
frame7.grid(row=1,column=1,sticky=N+S+E+W)
frame8 = Frame(master=window, width=400, height=height1, bg="#2E8BC0")
frame8.grid(row=2,column=1,rowspan=6,sticky=N+S+E+W)
frame9 = Frame(master=window, width=width1, height=height1, bg="#2E8BC0")
frame9.grid(row=4,column=0,sticky=N+S+E+W)





#Createing the button for prediction
btn_age = Button(
    window,
    text="Predict",
    font="Verdana 13 bold",
    command=handle_click_pediction,
    width=width,
    height=height,
    bg='#B1D4E0',
    relief=RIDGE,
    state=DISABLED
    )
btn_age.grid(row=1, column=0)

   


#Createing label for age
lbl_age = Label(
    window,
    text="Age:",
    font="Verdana 13 bold",
    fg='white',
    width=width3,
    height=height,
    bg='#2E8BC0'
    
    )
lbl_age.grid(row=2,column=0, sticky=W)

#Createing text box for age
text_age = Text(
    window,
    width=width4,
    height=height,
    font="Verdana 10",
    relief=RIDGE,
    state=DISABLED)


text_age.grid(row=2,column=0,sticky=E)

#Createing label for  gender
lbl_gender = Label(
    window,
    text="Gender:",
    font="Verdana 13 bold",
    fg='white',
    width=width3,
    height=height,
    bg='#2E8BC0')
lbl_gender.grid(row=3,column=0, sticky=W)

#Createing text box for gender
text_gender = Text(
    window,
    font="Verdana 10",
    width=width4,
    height=height,
    relief=RIDGE,
    state=DISABLED)

text_gender.grid(row=3,column=0,sticky=E)

#Createing label for ethnicity
lbl_ethnicity = Label(
    window,
    text="Ethnicity:",
    font="Verdana 13 bold",
    fg='white',
    width=width3,
    height=height,
    bg='#2E8BC0')
lbl_ethnicity.grid(row=4,column=0, sticky=W)

#Createing text box for ethnicity
text_ethnicity = Text(
    window,
    font="Verdana 10",
    width=width4,
    height=height,
    relief=RIDGE,
    state=DISABLED)

text_ethnicity.grid(row=4,column=0,sticky=E)

#Createing function to clear the text boxes
def clear_text():
    if (text_gender['state'] == DISABLED):
            text_gender['state'] = NORMAL
            
    if (text_age['state'] == DISABLED):
            text_age['state'] = NORMAL
            
    if (text_ethnicity['state'] == DISABLED):
            text_ethnicity['state'] = NORMAL
            
    text_gender.delete("1.0","end")
    text_age.delete("1.0","end")
    text_ethnicity.delete("1.0","end")
    
    text_age['state'] = DISABLED
    text_gender['state'] = DISABLED
    text_ethnicity['state'] = DISABLED

#Createing button for clearing the text boxes
btn_clear = Button(
    window,
    text="Clear text",
    font="Verdana 13 bold",
    width=width,
    height=height,
    bg='#B1D4E0',
    command=clear_text,
    relief=RIDGE)
btn_clear.grid(row=5,column=0)

#Createing an exit button
btn_exit = Button(
    window,
    text="EXIT",
    font="Verdana 13 bold",
    width=width,
    height=height,
    bg='#B1D4E0',
    command=window.destroy,
    relief=RIDGE)
btn_exit.grid(row=6,column=0)


#Createing a function to upload files from the computer
def open_file():
    global my_image_label
    global my_image
    btn_age['state'] = NORMAL
    window.filename = filedialog.askopenfilename(initialdir='This PC:',title='Select a file',filetypes=(('png files','*.png'),('jpg files','*.jpg')))
    img1=Image.open(window.filename)
    img1.thumbnail((250,250))
    my_image = ImageTk.PhotoImage(img1)
    my_image_label=Label(window,image=my_image,bg="#2E8BC0").grid(row=2,column=1,rowspan=6,sticky=N)

#Createing button for accesing files from the computer
btn_browse = Button(
    window,
    text='Open File',
    font="Verdana 13 bold",
    command=open_file,
    width=width,
    height=height,
    bg='#B1D4E0',
    relief=RIDGE)
btn_browse.grid(row=1,column=1)

window.mainloop()


    


```
